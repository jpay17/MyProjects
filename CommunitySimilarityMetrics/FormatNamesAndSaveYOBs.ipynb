```python
import csv
import networkx as nx
```

```python
with open('stn_phil_list2.csv', 'r') as infile:
    reader = csv.DictReader(infile)
    fieldnames = reader.fieldnames
    with open('phil_list_identifiers.csv', 'w') as outfile:
        fieldnames += ['identifier']
        writer = csv.DictWriter(outfile, fieldnames)
        writer.writeheader()
        for row in reader:
            name_parts = row['name'].split(' ')
            full_name = name_parts.pop()
            while name_parts:
                full_name += ' ' + name_parts.pop(0)[0]
            row['identifier'] = full_name.lower()
            writer.writerow(row)
    
#     writer = csv.DictWriter(f)
```

```python
node_attrs_dict = {}
with open('phil_list_identifiers.csv', 'r') as f:
    reader = csv.DictReader(f)
    for row in reader:
        node_attrs_dict[row['identifier']] = row['yob']
```

```python
import networkx as nx
g = nx.read_gpickle('citation_graph_pickle')
names = node_attrs_dict.keys()
nodes = g.nodes()
drop_list = []
for name in names:
    if name not in nodes:
        drop_list.append(name)
for name in drop_list:
    node_attrs_dict.pop(name, None)
nx.set_node_attributes(g, 'yob', node_attrs_dict)
```

```python
degs = g.degree()
nx.set_node_attributes(g, 'degree', degs)
```

```python
for n in [m for m in g.nodes(data=True) if 'canonical' not in m[1]]:
    n[1]['canonical'] = True
```

```python
canonical_canonical_nodes = [n[0] for n in g.nodes(data=True) 
                             if 'canonical' in n[1] and n[1]['canonical']
                             and n[1]['degree'] > 0]
zero_degree_nodes = [n[0] for n in g.nodes(data=True) if n[1]['degree'] == 0]
name_equivalences = []
uncaught_maybes = []
for n in zero_degree_nodes:
    
    matches = [cn for cn in canonical_canonical_nodes if n+' ' in cn]
    uncaught = [cn for cn in canonical_canonical_nodes if n+' ' not in cn and n in cn]
    for m in matches:
        name_equivalences.append((n, m))
    for u in uncaught:
        uncaught_maybes.append((n, u))
name_equivalences
for n in zero_degree_nodes:
    if n not in [neq[0] for neq in name_equivalences]:
        print(n)
print(len(name_equivalences), name_equivalences)
# canonical_canonical_nodes
with open('equivalent_names.csv', 'w') as f:
    fieldnames = ['name', 'canonical_name']
    writer = csv.DictWriter(f, fieldnames)
    writer.writeheader()
    for n in name_equivalences:
        writer.writerow({'name': n[0], 'canonical_name': n[1]})
        # print(uncaught_maybes) 
# for n in zero_degree_nodes:
#     h.remove_node(n)
# zero_degree_nodes = [n[0] for n in h.nodes(data=True) if n[1]['degree'] == 0]
```
